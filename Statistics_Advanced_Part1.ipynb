1. What is a random variable in probability theory?
    - A random variable is a function that assigns a numerical value to each outcome in a sample space of a random experiment.

2. What are the types of random variables?
   - There are two main types of random variables in probability theory:
     - Discrete Random Variable: Takes on a countable number of distinct values   e.g., number of heads in 3 coin tosses.
     - Continuous Random Variable: Takes on an infinite number of values within a range e.g., height of students, time taken.

3. What is the difference between discrete and continuous distributions?
   - The key difference between discrete and continuous distributions lies in the type of values the random variable can take:
     - Discrete distribution:
       Values are countable finite or countably infinite.
       Example: Number of dice rolls to get a 6, number of students.
     - Continuous distribution:
       Values form an unbroken interval, uncountably infinite.
       Example: Height, weight, temperature.

4. What are probability distribution functions (PDF)?
   - A Probability Distribution Function (PDF) describes how probabilities are distributed over the values of a continuous random variable.
   - The PDF, usually denoted as f(x), gives the relative likelihood of the     variable taking on a specific value.
   - For continuous variables,
     P(X=x)=0; instead, we compute P(a≤X≤b) by integrating the PDF over that interval.
   - The area under the entire PDF curve equals 1.

5.  How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
   - PDF for continuous variables: Describes the likelihood of the variable near a specific value. It is the derivative of the CDF.
   - CDF: Describes the probability that the variable is less than or equal to a value.
   - In short: PDF = rate of change of CDF, while CDF = accumulated probability up to x.

6. What is a discrete uniform distribution?
   - A discrete uniform distribution is a probability distribution where all outcomes are equally likely over a finite set of values.
   - Example:
     Rolling a fair 6-sided die → Each face (1–6) has probability 1/6.

7. What are the key properties of a Bernoulli distribution?
   - The Bernoulli distribution models a random experiment with only two possible outcomes: success (1) or failure (0).
   - Properties:
     - Outcomes: X=1 (success), X=0 (failure)
     - Probability of success: P(X=1)=p
     - Probability of failure: P(X=0)=1−p
     - Mean (Expected value): E[X]=p
     - Variance: Var(X)=p(1−p)

11. What are the characteristics of a normal distribution?
    - The normal distribution is a fundamental concept in probability and statistics, known for its distinctive bell-shaped curve. It models many naturally occurring variables, such as heights, test scores, and measurement errors.
    - It’s defined by two parameters: the mean (μ) and standard deviation (σ). The mean, median, and mode are all equal.
    - About 68% of values lie within 1σ of the mean, 95% within 2σ, and 99.7% within 3σ it is called Empirical Rule.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
   - Centeral Limit Theorem (CLT): When you take a large enough random sample from any population (with finite mean and variance), the sampling distribution of the sample mean will approximate a normal distribution, regardless of the population’s original distribution.
   - It is critical because:
     - Normality from any distribution: CLT allows us to use the normal distribution even if the population itself isn’t normally distributed.
     - Enables inference: It forms the basis for constructing confidence intervals and performing hypothesis tests using the normal or t-distribution.
     - Applies widely: Useful in fields like economics, psychology, biology, and machine learning where data often isn't normally distributed, but sample means still behave predictably.

  
14. How does the Central Limit Theorem relate to the normal distribution?
  - The Central Limit Theorem (CLT) explains why the normal distribution is so frequently used in statistics.
  - Shape of sampling distribution: CLT says that if you take repeated random samples of size n from any population (with finite mean and variance), the distribution of the sample means will approach a normal distribution as n becomes large even if the original population is not normal.
  - Practical implication: Because of CLT, we can apply techniques based on the normal distribution (like z-tests and confidence intervals) to problems involving averages or proportions, even when the data is not normally distributed.
  - Rule of thumb: A sample size of n≥30 is often sufficient for the CLT to hold.


15. What is the application of Z statistics in hypothesis testing?
    - Z-statistics or Z-tests are used in hypothesis testing when we want to determine whether a sample comes from a population with a known mean and standard deviation.
    - Applications in hypothesis testing:
      - Testing population mean: When the population standard deviation σ is known.
      - Comparing two means: To check if two populations have the same mean assuming known variances and large sample sizes.
      - Proportion tests: Testing if the proportion of success in a population equals a hypothesized value.

16. How do you calculate a Z-score, and what does it represent?
   - A Z-score measures how many standard deviations a data point is from the mean of a distribution.
     - Formula:    X = (X - μ)/σ
     - where:
       - X = individual data point
       - 𝜇 = population mean
       - 𝜎 = population standard deviation
    - It represents:
      - A Z-score of 0 means the value is exactly at the mean.
      - A positive Z-score means the value is above the mean.
      - A negative Z-score means the value is below the mean.
      - It helps compare scores across different distributions or assess how unusual a value is.

17. What are point estimates and interval estimates in statistics?
    - In statistics, point estimates and interval estimates are two ways to estimate unknown population parameters like mean or proportion based on sample data.
    - Point Estimate:A single value used as an estimate of a population parameter.
      - Example: The sample mean x is a point estimate of the population mean μ.
      - It's precise but gives no information about accuracy or uncertainty.
    - Interval Estimate: A range of values, usually given with a confidence level, that likely contains the population parameter.
      - Example: A 95% confidence interval for the mean might be [48.2, 51.6].
      - It expresses uncertainty and provides more useful information than a point estimate.
    

18. What is the significance of confidence intervals in statistical analysis?
   - Confidence intervals (CIs) are crucial in statistical analysis because they provide a range of plausible values for an unknown population parameter like mean or proportion, rather than just a single estimate.
   - Significance of Confidence Intervals:
     - Measure of uncertainty: A CI shows how much variability or uncertainty exists around a point estimate.
     - Informed decisions: Instead of guessing a single value, analysts can make more reliable decisions using the range.
     - Confidence level (e.g., 95%): Tells us that if we repeated the sampling many times, about 95% of the intervals would contain the true parameter.
     - Avoids binary thinking: Unlike hypothesis testing which often ends in “reject” or “don’t reject,” CIs give more nuanced information.
     - Used across fields: Widely applied in medicine, economics, engineering, and science for estimation and comparison.

20. How are Z-scores used to compare different distributions?
   - Z-scores allow comparison of values from different distributions by converting them into a standard scale, the standard normal distribution.
   - How they help:
      - Standardization: A Z-score transforms any value into a unit based on standard deviations from the mean: Z = (X - μ)/σ
      - Fair comparison: We can compare, for example, a student’s math score (mean = 70, σ = 10) with their history score (mean = 80, σ = 5), even though the scales are different.
      - Interpretability:A Z-score of 2 in any distribution means the value is 2 standard deviations above the mean.
      This makes it easy to judge relative performance or rarity of data points.

21. What are the assumptions for applying the Central Limit Theorem?
    - The Central Limit Theorem (CLT) works under a few important assumptions that ensure the sampling distribution of the sample mean becomes approximately normal:
    - Random Sampling: Data must be collected using random, independent samples from the population.
    - Independence: Each sample point should be independent of the others    especially true for simple random samples.
    - Sample Size (n):
       - The sample size should be large: n≥30 is a common rule of thumb.
       - For highly skewed or non-normal populations, a larger n may be needed.
    - Finite Variance: The population must have a finite standard deviation i.e., no infinite variability.
